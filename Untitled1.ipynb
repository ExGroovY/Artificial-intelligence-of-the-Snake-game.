```python
import pandas as pd

# Load the processed CSV
df = pd.read_csv("storm_data_2020_2025.csv")

# Drop unnecessary columns
df = df.drop(columns=["number", "step", "surface", "valid_time"])

# Define features (X)
X = df[["u10", "v10", "t2m", "sp", "msl"]]  # Input weather parameters

# Define target variable (Y) (Label storms based on pressure drops)
df["storm"] = (df["sp"] < 100000) & (df["msl"] < 101000)  # Low pressure = Storm

Y = df["storm"].astype(int)  # Convert to binary classification (0 = No Storm, 1 = Storm)
```

```python
from sklearn.model_selection import train_test_split

# Split data (80% training, 20% testing)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Train the Random Forest Model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, Y_train)

# Predict on test set
Y_pred = model.predict(X_test)

# Check accuracy
accuracy = accuracy_score(Y_test, Y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

```python
import numpy as np
import pandas as pd

# Generate 5 years (2026-2030) of synthetic weather data
num_days = 5 * 365  # 5 years of daily data

future_weather = pd.DataFrame({
    "u10": np.random.uniform(-10, 10, num_days),  # Wind speed U component
    "v10": np.random.uniform(-10, 10, num_days),  # Wind speed V component
    "t2m": np.random.uniform(270, 310, num_days),  # Temperature (Kelvin)
    "sp": np.random.uniform(90000, 105000, num_days),  # Surface pressure
    "msl": np.random.uniform(98000, 103000, num_days)  # Mean sea level pressure
})

# Add time column (dates from 2026 to 2030)
future_weather["date"] = pd.date_range(start="2026-01-01", periods=num_days, freq="D")

# Print sample data
print(future_weather.head())
```

```python
# Extract features for prediction
X_future = future_weather[["u10", "v10", "t2m", "sp", "msl"]]

# Predict storms using the trained model
future_weather["storm_prediction"] = model.predict(X_future)

# Print first 20 predictions
print(future_weather[["date", "storm_prediction"]].head(20))
```

```python
import matplotlib.pyplot as plt

storm_days = future_weather["storm_prediction"].sum()
normal_days = len(future_weather) - storm_days

plt.bar(["Storm Days", "Normal Days"], [storm_days, normal_days], color=["red", "blue"])
plt.title("Predicted Storm vs Normal Days (2026-2030)")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(future_weather["date"], future_weather["storm_prediction"], color="red")
plt.title("Storm Prediction Timeline (2026-2030)")
plt.xlabel("Year")
plt.ylabel("Storm Prediction (1 = Storm, 0 = No Storm)")
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd

# Convert the 'date' column to datetime format
df["date"] = pd.to_datetime(df["date"])

# Extract year and month for further analysis
df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
```

```python
# Count number of storm days per year
yearly_storms = df.groupby("year")["storm"].sum()

# Plot the number of storm days per year
plt.figure(figsize=(12, 5))
yearly_storms.plot(kind="bar", color="red", edgecolor="black")

# Labels & Formatting
plt.title("Total Storm Days Per Year (2020-2030)", fontsize=14, fontweight="bold")
plt.xlabel("Year", fontsize=12)
plt.ylabel("Number of Storm Days", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

```python
import seaborn as sns

# Pivot table to create a heatmap
storm_heatmap = df.pivot_table(values="storm", index="year", columns="month", aggfunc="sum")

# Create the heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(storm_heatmap, cmap="Reds", annot=True, fmt=".0f", linewidths=0.5)

# Labels & Formatting
plt.title("Monthly Storm Frequency (2020-2030)", fontsize=14, fontweight="bold")
plt.xlabel("Month", fontsize=12)
plt.ylabel("Year", fontsize=12)
plt.xticks(rotation=0)
plt.show()
```
